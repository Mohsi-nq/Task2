```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/kaggle/input/titanic-dataset/Titanic-Dataset.csv')
df.head()
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df['Age'].hist(bins=30)
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
numeric = df.select_dtypes(include=['int64', 'float64']).columns

df[numeric].hist(bins=30, figsize=(12, 8))
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(y=df['Age'])
plt.title('Boxplot of Age')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')

numeric_cols = titanic.select_dtypes(include='number').columns
for col in numeric_cols:
    sns.boxplot(y=titanic[col])
    plt.title(col)
    plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title('Age vs Survival')
plt.show()
```

```python
numeric_data =df.select_dtypes(include='number')
corr_matrix = numeric_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(numeric_data)
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']], hue='Survived')
```

```python
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
```

```python
sns.countplot(x='survived', data=titanic)
plt.title("Survival Count")
plt.show()

sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival by Gender")
plt.show()

sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title("Survival by Passenger Class")
plt.show()

sns.histplot(titanic['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

sns.boxplot(x=titanic['fare'])
plt.title("Fare Outliers")
plt.show()

sns.heatmap(titanic.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 10, 20, 40, 60, 80])
df.groupby('AgeGroup')['Survived'].mean()
```
